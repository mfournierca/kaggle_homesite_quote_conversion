```python
def clean_data(df):
    df["Original_Quote_Date"] = df["Original_Quote_Date"].apply(pd.to_datetime)
    df["Original_Quote_Year"] = df["Original_Quote_Date"].apply(lambda d: d.year)
    df["Original_Quote_Month"] = df["Original_Quote_Date"].apply(lambda d: d.month)
    df["Original_Quote_Weekday"] = df["Original_Quote_Date"].apply(lambda d: d.dayofweek)
    del df["Original_Quote_Date"]
    del df["QuoteNumber"]
    df = df.dropna(how="any", axis=1)
    return df
```

```python
df = clean_data(df)
```

```python
def get_categorical(df, col="GeographicField64"):
    df_cat = pd.merge(
        df[[c for c in df.columns if c != col]], 
        pd.get_dummies(df[col], prefix=col), 
        how="inner", 
        left_index=True, 
        right_index=True
    )
    return df_cat
```

```python
df.select_dtypes(exclude=["number"]).columns
```

```python
# split all columns with less than 100 unique values into categories

minuniq = 10
for col in df.columns:
    if col == "QuoteConversion_Flag":
        continue
    s = set()
    for val in df[col]:
        s.add(val)
        if len(s) > minuniq:
            break
    
    if len(s) <= minuniq or (df[col].dtype not in [np.dtype("float"), np.dtype("int")]):
        print("transforming to categorical: {}".format(col))
        df = get_categorical(df, col=col)
    else:
        print("skipping: {}".format(col))
        
df.head()
```

### Model Building

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df_varthreshold, 
    df["QuoteConversion_Flag"], 
    test_size=0.25, 
    random_state=3
)
```

```python
X_train.head()
```

#### Logistic Regression

##### build

```python
from sklearn.linear_model import LogisticRegressionCV

lr_model = LogisticRegressionCV(
    Cs=10,
    cv=10,
    scoring="log_loss"
)
lr_model.fit(X_train, y_train)
```

```python
lr_model.score(X_train, y_train)
```

##### evaluate

```python
lr_model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, lr_model.predict(X_test))
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, lr_model.predict_proba(X_test))
```
