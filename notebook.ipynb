## Clean Data

```python
def clean_data(df):
    df["Original_Quote_Date"] = df["Original_Quote_Date"].apply(pd.to_datetime)
    df["Original_Quote_Month"] = df["Original_Quote_Date"].apply(lambda d: d.month)
    df = df.dropna(how="any", axis=1)
    return df
```

```python
def get_categorical(df, col="GeographicField64"):
    df_region = pd.merge(
        df, 
        pd.get_dummies(df[col]), 
        how="inner", 
        left_index=True, 
        right_index=True
    )
    return df_region
```

### Model Building

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df_varthreshold, 
    df["QuoteConversion_Flag"], 
    test_size=0.33, 
    random_state=1
)
```

#### Logistic Regression

##### build

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(
    X_train,
    y_train
)
```

##### evaluate

```python
lr_model.score(X_test, y_test)
```

#### Random Forest Classifier

##### build

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(
    X_train, 
    y_train
)
```

##### evaluate

```python
rf_model.score(
    X_test,
    y_test
)
```
